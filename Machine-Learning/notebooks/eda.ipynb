```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw.csv")
```

```python
pd.set_option("display.max_columns", None)
```

```python
fig, ax = plt.subplots(3, 2, figsize=(16, 18))


# Hist
sns.histplot(data=df["Max Loan Amount (USD)"], ax=ax[0, 0], color='steelblue')
ax[0, 0].set_title("Max Loan Amount Hist")

# Boxplot
sns.boxplot(y="Max Loan Amount (USD)", data=df, ax=ax[0, 1], color='steelblue')
ax[0, 1].set_title("Max Loan Amount Boxplot")

# Scatterplots
sns.scatterplot(x="Annual Income (USD)", y="Max Loan Amount (USD)", data=df, ax=ax[1, 0], color='steelblue', alpha=0.3)
ax[1, 0].set_title("Annual Income Scatter Plot")

sns.scatterplot(x="Credit Score", y="Max Loan Amount (USD)", data=df, ax=ax[1, 1], color='steelblue', alpha=0.3)
ax[1, 1].set_title("Credit Score Scatter Plot")

sns.scatterplot(x="Down Payment (USD)", y="Max Loan Amount (USD)", data=df, ax=ax[2, 0], color='steelblue', alpha=0.3)
ax[2, 0].set_title("Down Payment Scatter Plot")

sns.scatterplot(x="Age", y="Max Loan Amount (USD)", data=df, ax=ax[2, 1], color='steelblue', alpha=0.3)
ax[2, 1].set_title("Age Scatter Plot")


# Heatmap
fig0, ax0 = plt.subplots(figsize=(8, 5))

corr = df[
    ["Max Loan Amount (USD)", "Interest Rate", "Down Payment (USD)", "Credit Score", "Existing Monthly Debt (USD)", "Loans Repaid", "Annual Income (USD)", "Employment Years"]
].corr()
sns.heatmap(data=corr, annot=True, ax=ax0, color='steelblue')
ax0.set_title("Correlation Matrix")


plt.tight_layout()
plt.show()
```
